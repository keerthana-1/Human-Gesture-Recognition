```python
import pandas as pd
from scipy.io import arff
import os

data_path = "/content/drive/MyDrive/NATOPS/"
output_csv = "multidimensional_timeseries_data.csv"
```

```python
train_data = []

final_df=pd.DataFrame()

for i in range(1,25):
    filename = f"NATOPSDimension{i}_TRAIN.arff"
    filepath = os.path.join(data_path, filename)

    result_df=pd.DataFrame()
    #reading data from .arff file
    df_dimension, meta = arff.loadarff(open(filepath, 'r'))

    #converting the data to dataframe
    train_df = pd.DataFrame(df_dimension)
    values_list=[]

    for j in range(51):
      colname='channel_'+str(i-1)+'_'+str(j)
      values_list.extend(train_df[colname].tolist())

    result_df[f'dimension_{i}'] = values_list

    final_df=pd.concat([final_df, result_df],axis=1)

final_df = final_df.assign(ClassAttribute=train_df['classAttribute'].tolist() * 51)
final_df = final_df.assign(test_train='train')
```

```python
final_df.head()
```

```python
final_df1=pd.DataFrame()

for i in range(1,25):
    filename = f"NATOPSDimension{i}_TEST.arff"
    filepath = os.path.join(data_path, filename)

    result_df=pd.DataFrame()
    #reading data from .arff file
    df_dimension, meta = arff.loadarff(open(filepath, 'r'))

    #converting the data to dataframe
    test_df = pd.DataFrame(df_dimension)
    values_list=[]

    for j in range(51):
      colname='channel_'+str(i-1)+'_'+str(j)
      values_list.extend(test_df[colname].tolist())

    result_df[f'dimension_{i}'] = values_list

    final_df1=pd.concat([final_df1, result_df],axis=1)

final_df1 = final_df1.assign(ClassAttribute=test_df['classAttribute'].tolist() * 51)
final_df1 = final_df1.assign(test_train='test')
```

```python
final_df1.head()
```

```python
result = pd.concat([final_df, final_df1], ignore_index=True)
```

```python
result['id'] = result.reset_index().index
result = result[['id'] + [col for col in result.columns if col != 'id']]
```
